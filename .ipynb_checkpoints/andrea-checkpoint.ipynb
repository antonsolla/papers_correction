```python
def basic_cleaning(text):
    # Convert to string
    text = str(text).strip().lower()
    # Replace digits with words
    text = ''.join([num2words(char) if char.isdigit() else char for char in text])
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

# Apply the cleaning function
df['text1'] = df['text1'].apply(basic_cleaning)
df['text2'] = df['text2'].apply(basic_cleaning)


df.head()
```

```python
# Check if the function remove numbers worked
contains_digits1 = df['text1'].apply(lambda x: any(char.isdigit() for char in x))
contains_digits1.sum()
```

```python
# Check if the function remove numbers worked
contains_digits2 = df['text2'].apply(lambda x: any(char.isdigit() for char in x))
contains_digits2.sum()
```

```python
import re

def regex(text):
    if isinstance(text, list):
        return [re.sub('<[^<]+?->', '', word) for word in text]
    else:
        return re.sub('<[^<]+?->', '', text)

# Tokenized  all rows in both columns
df['text1']= df['text1'].apply(regex)
df['text2']= df['text2'].apply(regex)

df.head()
```

```python
#tokenized the text
from nltk.tokenize import word_tokenize

df['text1_tokenized'] = (df['text1']).apply(word_tokenize)
df
```

```python
df['text2_tokenized'] = (df['text2']).apply(word_tokenize)
df
```

```python
from nltk.stem import WordNetLemmatizer

#Create a function to Lemmatize
def lemma(text):
    lemmatizer = WordNetLemmatizer() # Instantiate lemmatizer
    lemmatized = [lemmatizer.lemmatize(word) for word in text] # Lemmatize
    lemmatized_string = " ".join(lemmatized)
    return lemmatized_string

df['text1_tokenized'] = df['text1_tokenized'].apply(lemma)
df['text2_tokenized'] = df['text2_tokenized'].apply(lemma)

df.head()
```

```python
text_combined = df['combined']
text_combined
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(text_combined)
```

```python
from scipy.cluster.hierarchy import ward, dendrogram
import matplotlib.pyplot as plt

# Use the ward clustering method; it minimizes the variance of clusters being merged.
linkage_matrix = ward(tfidf_matrix.toarray())

# Plot the dendrogram
plt.figure(figsize=(8, 4))
dendrogram(linkage_matrix, labels=[f"Document {i+1}" for i in range(len(text_combined))], leaf_rotation=90, leaf_font_size=12)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Document')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()
```
